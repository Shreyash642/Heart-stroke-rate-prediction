```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('healthcare-dataset-stroke-data.csv')
df
```

```python
df.isnull().sum()
```

```python
df['bmi'].fillna(df['bmi'].median(),inplace=True)
df.isnull().sum()
```

```python
df['work_type'].value_counts()
```

```python
df['work_type'].replace('Private',0,inplace=True)
df['work_type'].replace('Self-employed',1,inplace=True)
df['work_type'].replace('children',2,inplace=True)
df['work_type'].replace('Govt_job',3,inplace=True)
df['work_type'].replace('Never_worked',4,inplace=True)
```

```python
df['ever_married'].value_counts()
```

```python
df['ever_married'].replace('Yes',0,inplace=True)
df['ever_married'].replace('No',1,inplace=True)
df
```

```python
df['Residence_type'].value_counts()
```

```python
df['Residence_type'].replace('Urban',0,inplace=True)
df['Residence_type'].replace('Rural',1,inplace=True)
df
```

```python
df['smoking_status'].value_counts()
```

```python
df['smoking_status'].replace('never smoked',0,inplace=True)
df['smoking_status'].replace('Unknown',1,inplace=True)
df['smoking_status'].replace('formerly smoked',2,inplace=True)
df['smoking_status'].replace('smokes',3,inplace=True)
df
```

```python
df['gender'].value_counts()
```

```python
df['gender'].replace({'Male':0,'Female':1,'Other':2},inplace=True)
df
```

```python
df.head()
df['gender'].astype(int)
df.dtypes
```

# KNN

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=df.drop(['stroke'],axis=1)
X
```

```python
y=df['stroke']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standardizes the data 
scaler=StandardScaler()
scaler.fit(X)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# initializes and runs the classifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
```

```python
# gives the confusin matrix
conf=metrics.confusion_matrix(y_test,y_pred)
print(conf)
```

```python
print(metrics.classification_report(y_test,y_pred))
```

# Training using K-nn

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
parameters={'n_neighbors':[3,5,11,19],
           'weights':['uniform','distance'],
           'metric':['minikowski','manhattan']}
```

```python
clf=GridSearchCV(KNeighborsClassifier(),parameters,cv=3,verbose=1,n_jobs=-1)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
clf=KNeighborsClassifier(metric='manhattan',n_neighbors=11,weights='uniform')
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report 
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
